```python
"""数据读取"""
import pandas as pd
import numpy as np
import networkx as nx
province = pd.read_csv('data/nmProv.csv',header=None)[0].tolist()
industry = pd.read_csv('data/nmInd.csv',header=None)[0].tolist()
names = []
for provin in province:
    for ind in industry:
        names.append(provin+ind)
data  = pd.read_csv('data/inoutput2017.csv', header=None, names=names)
for i in range(data.shape[0]):
        data.iloc[i,i]=0
# data = data.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
for i in range(data.shape[0]):
        data.iloc[i,i]=1
data.to_csv('data/adj_matrix.csv',index=False)
data
```

```python
array_data = np.array(data)
G = nx.Graph(array_data)
pagerank = nx.pagerank(G)
list1= sorted(pagerank.items(),key=lambda x:x[1],reverse=True)
list1
pagerank_sorted_name = list()
pagerank_sorted_value = list()
for one_tuple in list1:
    pagerank_sorted_name.append(names[one_tuple[0]])
    pagerank_sorted_value.append(one_tuple[1])
pagerank_result = dict()
pagerank_result['province_industry'] = pagerank_sorted_name
pagerank_result['pagerank_value'] = pagerank_sorted_value
pagerank_result = pd.DataFrame(pagerank_result)
pagerank_result.to_csv('result/pagerank_result.csv',index=False)
```

```python
hits = nx.hits(G)
```

```python
hits_result = dict()
hits_result['province_industry'] = names
hits_result['hubs'] = list(hits[0].values())
hits_result['authority'] = list(hits[1].values())
hits_result = pd.DataFrame(hits_result)
hits_result.to_csv('result/hits_result.csv',index=False)
```
