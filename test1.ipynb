```python
"""试图使用py把.mat文件转化成为.csv文件"""
import pandas as pd
import numpy as np
inoutput2017  = pd.read_csv('data/inoutput2017.csv', header=None, names=["col_{}".format(i) for i in range(1395)])
inoutput2017
```

```python
for i in range(inoutput2017.shape[0]):
    inoutput2017.iloc[i,i]=0
inoutput2017
```

```python
num_industry = 45
name = pd.read_csv("data/nmInd.csv",header=None)
name = name[0].tolist()
pi = 0
start = pi * num_industry
end = (1+pi) * num_industry
data_index = list(range(start, end))
tnet = inoutput2017.iloc[data_index,data_index]
tnet
```

```python
import networkx as nx
tg = nx.DiGraph()
tg.add_nodes_from(name)
for i in range(45):
    for j in range(45):
        start_node = name[i]
        end_node = name[j]
        weight = tnet.iloc[i,j]
        tg.add_edge(start_node, end_node,weight=weight)
```

```python
e = len(list(tg.edges))
vn = 8
n= len(list(tg.nodes))
es = np.ones((e,1))
vs = np.zeros((n,3))
nx = 10
lv = len(vn)
```
