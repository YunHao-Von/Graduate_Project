```python
"""数据读取"""
import pandas as pd
import numpy as np
import networkx as nx
province = pd.read_csv('data/nmProv.csv',header=None)[0].tolist()
industry = pd.read_csv('data/nmInd.csv',header=None)[0].tolist()
names = []
for provin in province:
    for ind in industry:
        names.append(provin+ind)
data  = pd.read_csv('data/inoutput2017.csv', header=None, names=names)
for i in range(data.shape[0]):
        data.iloc[i,i]=1
data = data.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)),axis=1)
for i in range(data.shape[0]):
        data.iloc[i,i]=1
data.to_csv('data/adj_matrix.csv',index=False)
data
```

```python
from torch.utils.data import DataLoader
train_data = np.array(data)
train_loader = DataLoader(train_data,batch_size=128,shuffle=True)
for batch_idx,temp_data in enumerate(train_loader):
    print(temp_data.shape)
```

```python
import torch
from torch import nn
import torch.nn.functional as F
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
torch.set_default_tensor_type(torch.DoubleTensor)
device = torch.device('cuda:0' if torch.cuda.is_available() else "cpu")
class VAE(nn.Module):
    def __init__(self,hidden_1=256,hidden_2=256,in_decode_dim=2,hidden_3=256):
        super(VAE,self).__init__()
        self.fc1 = nn.Linear(1395,hidden_1)
        self.fc21 = nn.Linear(hidden_2,2)
        self.fc22 = nn.Linear(hidden_2,2)
        self.fc3 = nn.Linear(in_decode_dim,hidden_3)
        self.fc4 = nn.Linear(hidden_3,1395)
    
    def encode(self,x):
        h1 = F.relu(self.fc1(x))
        return self.fc21(h1),self.fc22(h1)
    
    def reparametrize(self,mean,lg_var):
        std = lg_var.exp().sqrt()
        eps = torch.FloatTensor(std.size()).normal_().to(device)
        return eps.mul(std).add_(mean)
    
    def decode(self,z):
        h3 = F.relu(self.fc3(z))
        return self.fc4(h3)
    
    def forward(self,x,*arg):
        mean,lg_var = self.encode(x)
        z = self.reparametrize(mean,lg_var)
        return self.decode(z),mean,lg_var
    
```

```python
reconstruction_function = nn.MSELoss(size_average=False)

def loss_function(recon_x, x, mean, lg_var):
    MSEloss = reconstruction_function(recon_x, x)  # mse loss
    KLD = -0.5 * torch.sum(1 + lg_var -  mean.pow(2) - lg_var.exp())
    return MSEloss*0.5 + KLD
```

```python
def train(model,num_epochs=270):
    optimizer = torch.optim.Adam(model.parameters(),lr=1e-3)
    display_step = 5
    loss_list = list()
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for batch_idx,batch_data in enumerate(train_loader):
            optimizer.zero_grad()
            batch_data = batch_data.to(device)
            recon_batch, mean, lg_var = model(batch_data)
            loss = loss_function(recon_batch,batch_data,mean,lg_var)
            loss.backward()
            train_loss += loss.data
            optimizer.step()
        if epoch % display_step == 0:
            print('Epoch:','%04d' % (epoch+1),'loss=',loss.data)
        loss_list.append(float(loss.data))
    return loss_list
    
            
```

```python
model = VAE().to(device)
loss_list =train(model,270)
```

```python
import matplotlib.pyplot as plt
plt.plot(loss_list)
plt.show()
```
